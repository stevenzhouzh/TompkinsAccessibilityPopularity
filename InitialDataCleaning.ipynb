```python
import numpy as np
import pandas as pd
```

```python
attractions_df = pd.read_csv("Attractions.csv")
food_df = pd.read_csv("Food_Drinks.csv")
groceries_df = pd.read_csv("Groceries_ConvinienceStores.csv")
shopping_df = pd.read_csv("Shopping.csv")

attractions_df.head()
```

```python
poi_df = pd.concat([attractions_df, food_df, groceries_df, shopping_df],ignore_index=True)
poi_df
```

```python
poi_df.columns
```

```python
poi_df = poi_df.drop(['openingHours/0/day', 'openingHours/0/hours', 'openingHours/1/day',
       'openingHours/1/hours', 'openingHours/2/day', 'openingHours/2/hours',
       'openingHours/3/day', 'openingHours/3/hours', 'openingHours/4/day',
       'openingHours/4/hours', 'openingHours/5/day', 'openingHours/5/hours',
       'openingHours/6/day', 'openingHours/6/hours','price', 'imagesCount', 'isAdvertisement',
       'description', 'rank', 'street', 'city', 'state'], axis=1)
```

```python
poi_df = poi_df.drop_duplicates(subset=None, ignore_index=True)
poi_df
```

```python
poi_df.to_csv("Tompkins_POI.csv", index=False)
```
